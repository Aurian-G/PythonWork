```python
#Simple Calculator
#Python v3.7.3
#12/13/19

print('''Welcome to my simple Calculator.
You will be asked to input two numbers and an operation.\n\n''')

quit_app = ""

while quit_app != "n":
    
    var1 = float(input("Please input the first number: "))
    
    print('''\n
    A - add\n
    B - subtract\n
    C - multiply\n
    D - divide''')
    
    opvar = str(input("Please pick which letter you want for your operation: "))
    var2 = float(input("Please input the second number: "))
    
    if opvar.lower() == "a":
        result = float(var1+var2)
        print(f"your answer is {result}")
    
    if opvar.lower() == "b":
        result = float(var1-var2)
        print(f"your answer is {result}")
        
    if opvar.lower() == "c":
        result = float(var1*var2)
        print(f"your answer is {result}")
    
    if opvar.lower() == "d":
        result = float(var1/var2)
        print(f"your answer is {result}")
    
    replay_calc = input("Please enter 'y' or 'n' if you would like to do another calculation.")
    
    if replay_calc == "n":
        print("Thanks for playing goodbye!")
        quit_app = "n"
```

```python
#Simple Calculator
#Python v3.7.3
#12/16/19

def display_ops(): #display operators
    
    print('''\n
    A - add\n
    B - subtract\n
    C - multiply\n
    D - divide''')
    
    while True:
        
        opvar = input("Please pick which letter you want for your operation: ")
        opvar1 = opvar.lower()
        
        if opvar1 == "a" or opvar1 == "b" or opvar1 == "c" or opvar1 == "d":
            var_list.append(opvar1)
            break
        else:
            print("Incorrect input, please try again!")
            continue

def var_check(): #integer check
    
    while True:
        try:
            var1 = float(input("Please enter a number:"))
            var_list.append(var1)
            break
        except ValueError:
            print("That is an invalid number. Please try again!")

print('''Welcome to my simple Calculator.
You will be asked to input a number, an operator, followed by another number.\n\n''')

def add_op(): #addition
    
    result = float(var1+var2)
    print(f"your answer is {result}")

def sub_op(): #subtraction
    
    result = float(var1-var2)
    print(f"your answer is {result}")

def mult_op(): #multiplication
    
    result = float(var1*var2)
    print(f"your answer is {result}")

def div_op(): #division
    
    result = float(var1/var2)
    print(f"your answer is {result}")

var_list = []
quit_app = ""

while quit_app != "n":
    
    var_check() #user enters 1st value
    display_ops() #user enters operator
    var_check() #user enters 2nd value
    
    var1 = var_list.pop(0)
    var2 = var_list.pop(1)
    opvar = var_list.pop(0)
    
    if opvar.lower() == "a":
        add_op()
    
    if opvar.lower() == "b":
        sub_op()
        
    if opvar.lower() == "c":
        mult_op()
    
    if opvar.lower() == "d":
        div_op()
    
    replay_calc = input("Please enter 'y' or 'n' if you would like to do another calculation.") #reset Calculator
    
    if replay_calc.lower() == "n":
        print("Thanks for playing goodbye!")
        quit_app = "n"
```
